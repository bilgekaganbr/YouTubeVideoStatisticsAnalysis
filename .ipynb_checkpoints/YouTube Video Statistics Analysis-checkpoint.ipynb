```python
def tag_count(tags):
    
    tagList = tags.split("|")
    
    return len(tagList)

youtube["tag_number"] = youtube["tags"].apply(tag_count) # Calculate the number of tags used in each video and add the resulting values as a new column

youtube
```

```python
# Sort videos from most to least liked

def likesdislikesratio(likes, dislikes):
    
    likesList = []
    
    for key, value in list(likes.iteritems()):
        
        likesList.append(value)
        
    dislikesList = []
    
    for key, value in list(dislikes.iteritems()):
        
        dislikesList.append(value)
        
    likedislike = list(zip(likesList, dislikesList))
    
    ratio = []
    
    for like, dislike in likedislike:
        
        if like + dislike == 0:
            ratio.append(0)
            
        else:
        
            ratio.append(like / (like + dislike))
    
    return ratio
    
youtube["likesdislikes_ratio"] = likesdislikesratio(youtube["likes"], youtube["dislikes"])

youtube.sort_values("likesdislikes_ratio", ascending = False, inplace = True)

youtube
```
